```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
bank = pd.read_csv('C:/Users/Lenovo/OneDrive/Desktop/assignments/bank-full.csv')
bank
```

```python
bank.info()
```

```python
data = pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome','month'])
data
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data
```

```python
data.info()
```

```python
#custom binary encoding of binary o/p variables
data['default'] = np.where(data['default'].str.contains('yes'),1,0)
data['housing'] = np.where(data['housing'].str.contains('yes'),1,0)
data['loan'] = np.where(data['loan'].str.contains('yes'),1,0)
data['Target'] = np.where(data['Target'].str.contains('yes'),1,0)

data
```

```python
data.info()
```

MODEL BUILDING

```python
#dividing data into input and output
x = pd.concat([data.iloc[:,:0:10],data.iloc[:,11:]],axis=1)
y = data.iloc[:,10]
```

```python
classifier = LogisticRegression()
classifier.fit(x,y)
```

MODEL PREDICTION

```python
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({'ACTUAL_Y ':y , 'Y_PRED_PROB':y_pred})
y_pred_df
```

TESTING MODEL ACCURACY

```python
#WE ARE TAKING CONFUSION MATRIX FOR MODEL ACCURACY
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
#the model accuracy is calculated by (a+d)/(a+b+c+d)
print(((39420+921)/(39420+502+4368+921))*100,'%')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

```python
classifier.predict_proba(x)[:,1]
```

```python
#roc curve plotting and finding the auc value
fpr,tpr,thresholds = roc_curve(y,classifier.predict_proba(x)[:,1])
auc = roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color = 'lime',label = 'logit model(area = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('FALSE POSITIVE RATE OR [1 - TRUE NEGATIVE RATE]')
plt.ylabel('TRUE POSITIVE RATE')
plt.show()

print('\nAUC ACCURACY: ',auc,'%')
```
