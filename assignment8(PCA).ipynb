PCA IMPLEMENTATION

```python
#Applying pca fit to the dataset
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
#PCA component matrix
pca.components_
```

```python
#The amount of varience that each pca has
var = pca.explained_variance_ratio_
var
```

```python
#CUMULATIVE VARIENCE of each PCA
import numpy as np
cum_var = np.cumsum(np.round(var,decimals=4)*100) 
cum_var
```

```python
import matplotlib.pyplot as plt
plt.plot(cum_var,color = 'black')
```

```python
#final dataframe
final_df = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#VISUALISATIONS
import seaborn as sns
fig = plt.figure(figsize=(16,12))
sns.scatterplot(data = final_df)
```

CHECKING WITH OTHER ALGORITHM

• HIERARCHIAL CLUSTERING

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,7))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
#creating clusters
hclusters = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean',linkage = 'ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine_norm),columns = ['Clustersid'])
y['Clustersid'].value_counts()
```

```python
WINE['Clustersid'] = hclusters.labels_
WINE
```

K MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

```python
# within cluster sum of squares criterian
wcss = []
for i in range(1,6):
    kmeans = KMeans(n_clusters = i,random_state = 2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
#plot k values range vs wcss to get elbow graph for choosing k (no. of clusteing)
plt.plot(range(1,6),wcss)
plt.title('ELBOW GRAPH')
plt.xlabel('NUMBER OF CLUSTERS')
plt.ylabel('WCSS')
```
